```python
#---Example: Manipulating DataFrame---
import pandas as pd
df = pd.read_csv("presidents.csv")
df.head() #list of US presidents from wikipedia
```

```python
#we'll start cleaning
#firtst, the last name and first name will be cleaned with regex
df["First"] = df['President']
#we replace the pattern that matches a last name
df["First"]=df["First"].replace("[ ].*", "", regex=True) #indicating the string after whitespace
df.head()
```

```python
#it's slow, not that efficient, but it works
#an alternative is to use the apply() function
#first, we'll drop the column

del(df["First"])
#the apply funtion on a dataframe will take some arbitray function and apply it to either a series or dataframe
#splitting a string into two pieces using a single row of data
def splitname(row):
    #extractint the firstname and creating a new entry in the series
    row['First']=row['President'].split(" ")[0]
    #same thing but with the last word in the string
    row['Last']=row['President'].split(" ")[-1]
    #we return the row and the pandas .apply() function will merge them back into the dataframe
    return row
#we apply this to the dataframe indicating we want to apply itr across columns
df=df.apply(splitname, axis='columns')
df.head()
```

```python
#it achieves the result
```

```python
#.extract() function
#first, we'll drop our firstname and lastname
del(df['First'])
del(df['Last'])

#we'll use regex as an input, we want to return groups and just have the firstname and lastname
#a solution is we match three groups but only return two, the first and last name
pattern="(^[\w]*)(?:.* )([\w]*$)"
        #first parameter indicates any character
        #second parameter ?: indicates that this group shouldn't be returned, we want any characters followed by a whitespace
        #last parameter is any characters but to anchor them to the back of the string
#the extract function is built into the str attribute of the series object
#we call it with Series.str.extract(pattern)
df["President"].str.extract(pattern).head()
```

```python
#naming the groups
pattern="(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"
#we call extract
names=df["President"].str.extract(pattern).head()
names
```

```python
#we can copy these intro our main dataframe
df["First"]=names["First"]
df["Last"]=names["Last"]
df.head()
```

```python
#now we'll clean the born column
#we will first get rid of anything that isn't in the pattern of month day and year
df["Born"]=df["Born"].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df["Born"].head()
```

```python
#we now cleaned up the date format, this column is an object
#pandas was made to deal with financial transactions, so it has very interesting function
df["Born"]=pd.to_datetime(df["Born"])
df["Born"].head()
```
