```python
#---Basic Statistical Testing---
#we're going to study hypothesis testing, it's goal is to determine if, for instance, the two different conditions we have 
#in an expertiment have resulted in different impacs.
import numpy as np
import pandas as pd

from scipy import stats
```

```python
#hyposthesis testing has two statements of interest:
#the first is out actual explanation, which we call the alternative hypothesis, 
#and the second is that the explanation we have is not sufficient, and we call this the null hypothesis.

#our actual testing method is to determine whether the null hypothesis is true ot not. 
#if we find that there is a difference between groups, then we can reject the null hypothesis and we accept out alternative.

df = pd.read_csv('grades.csv')
df.head()
```

```python
#let's look at some summary statistics
print("There are {} rows and {} columns".format(df.shape[0], df.shape[1]))
```

```python
#we'll segment the population into two pieces:
#early finishers, the ones that finish the first assignement by the end of December 2015
#late finishers, the ones that finish it after that

early_finishers=df[pd.to_datetime(df['assignment1_submission']) < '2016']
early_finishers.head()
```

```python
late_finishers=df[~df.index.isin(early_finishers.index)] #this is a great method to get the opposite of a dataframe
late_finishers.head()
```

```python
#we can see the means of the assignements calculates, let's compare
print(early_finishers['assignment1_grade'].mean())
print(late_finishers['assignment1_grade'].mean())
```

```python
#when doing hypothesis testing, we have to choose a significance level as a threshold for how much of a chance we're willing to accept.
#in this example, we'll use a theshold of 0.05 (alpha = 5%).

#students t-testing allows us to form the alternative hypothesis (different) as well as the null hypothesis (same)
from scipy.stats import ttest_ind

ttest_ind(early_finishers['assignment1_grade'], late_finishers['assignment1_grade'])
```

```python
#the probability is 0.18, this is above the alpha value of 0.5
#this means that we cannot reject the null hypothesis, where two are the same
```

```python
#other assignement grades
print(ttest_ind(early_finishers['assignment2_grade'], late_finishers['assignment2_grade']))
print(ttest_ind(early_finishers['assignment3_grade'], late_finishers['assignment3_grade']))
print(ttest_ind(early_finishers['assignment4_grade'], late_finishers['assignment4_grade']))
print(ttest_ind(early_finishers['assignment5_grade'], late_finishers['assignment5_grade']))
print(ttest_ind(early_finishers['assignment6_grade'], late_finishers['assignment6_grade']))
```

```python
#example of running p-values with more tests to get a value which is statistically significant just by chance
df1=pd.DataFrame([np.random.random(100) for x in range(100)]) #this is a list comprehension
df1.head()
```

```python
#a second dataframe
df2=pd.DataFrame([np.random.random(100) for x in range(100)])
```

```python
#alpha of 10%. we're going to compare each column in df1 to the same in df2 and report if it's less than 10%, which means we 
#have sufficient evidence to say that they are different

def test_columns(alpha=0.1):
    #how many differ
    num_diff=0
    #iterating over the columns
    for col in df1.columns:
        #run the test between  the two dataframes
        teststat,pval=ttest_ind(df1[col],df2[col])
        #checking pvalue vs alpha
        if pval<=alpha:
            #print out if they are different and increment the num_diff
            print("Col {} is statistically significantly different at alpha={}, pval={}".format(col, alpha, pval))
            num_diff=num_diff+1
    #printing some summary stats
    print("Total number different was {}, which is {}%".format(num_diff, float(num_diff)/len(df1.columns)*100))

test_columns()
```

```python
#testing other alpha values
test_columns(0.05)
```
